# Exploratory Data Analysis for AI-Powered Plant Disease Detection

This notebook is used for analyzing the dataset of plant leaf images. It includes visualizations and insights into the data distribution, which will help in understanding the characteristics of the dataset before training the model.

```python
# Import necessary libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image

# Set the path to the dataset
raw_data_path = '../data/raw/'

# Function to load images and their labels
def load_images_from_folder(folder):
    images = []
    labels = []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        if filename.endswith('.jpg') or filename.endswith('.png'):
            img = Image.open(img_path)
            images.append(img)
            labels.append(filename.split('_')[0])  # Assuming label is part of the filename
    return images, labels

# Load images
images, labels = load_images_from_folder(raw_data_path)

# Convert labels to a DataFrame
label_df = pd.DataFrame(labels, columns=['Label'])

# Visualize the distribution of labels
plt.figure(figsize=(10, 6))
sns.countplot(y='Label', data=label_df, order=label_df['Label'].value_counts().index)
plt.title('Distribution of Plant Disease Labels')
plt.xlabel('Count')
plt.ylabel('Label')
plt.show()

# Display some sample images
plt.figure(figsize=(12, 8))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(images[i])
    plt.title(labels[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
